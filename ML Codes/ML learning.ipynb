```python
y = df['Survived']
```

```python
X = df[df.columns[1:]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression(random_state =100,C=1.2, penalty='l2')
```

```python
model1.fit(X_train,y_train)
```

```python
y_predTr = model1.predict(X_train)
y_predTe = model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_train,y_predTr))
c = classification_report(y_train,y_predTr)
print(c)
```

```python
print(confusion_matrix(y_test,y_predTe))
c = classification_report(y_test,y_predTe)
print(c)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn import tree
```

```python
model2 = DecisionTreeClassifier(criterion='entropy')

model2.fit(X_train,y_train)
y_predTr = model2.predict(X_train)
y_predTe = model2.predict(X_test)
print(confusion_matrix(y_train,y_predTr))
c = classification_report(y_train,y_predTr)
print(c)
print(30*'-')
print(confusion_matrix(y_test,y_predTe))
c = classification_report(y_test,y_predTe)
print(c)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 20))
tree.plot_tree(model2, feature_names=X.columns, class_names='Survived', filled = True)
plt.savefig('my.pdf')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3 = RandomForestClassifier(n_estimators=100)
model3.fit(X_train,y_train)
y_predTr = model3.predict(X_train)
y_predTe = model3.predict(X_test)
print(confusion_matrix(y_train,y_predTr))
c = classification_report(y_train,y_predTr)
print(c)
print(30*'-')
print(confusion_matrix(y_test,y_predTe))
c = classification_report(y_test,y_predTe)
print(c)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
mod_lis = [model1,model2,model3]
```

```python
def training_resyl(li):
    for model in li:
        model = RandomForestClassifier(n_estimators=100)
        model.fit(X_train,y_train)
        y_predTr = model.predict(X_train)
        y_predTe = model.predict(X_test)
        print(confusion_matrix(y_train,y_predTr))
        c = classification_report(y_train,y_predTr)
        print(c)
        print(10*'-')
        print(confusion_matrix(y_test,y_predTe))
        c = classification_report(y_test,y_predTe)
        print(c)
        plot_confusion_matrix(model, X_test, y_test)  
        plt.show()
        
        print(40*'*-*')
```

```python
training_resyl(mod_lis)
```
