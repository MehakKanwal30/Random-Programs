```python
import random
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
lenght = 100
no_clusters = 3
no_features = 3
```

```python
features = []
for i in range(no_features):
    features.append([])

for i in range(no_features):
    features[i] = [random.randint(1, 500) for i in range(lenght)]
```

```python
c = []

for i in range(no_clusters):
    a = random.randint(0,lenght-1)
    
    feat = [features[j][a] for j in range(no_features)]
    c.append(feat)
```

```python
features
```

```python
def min_index(arr):
    minimum = min(arr)
    minimum_index = arr.index(minimum)
    return minimum_index
```

```python
cluster = []
for i in range(no_clusters):
    cluster.append([])
    
for i in range(lenght):
    f = []
    for j in range(no_clusters):
        temp2 = []
        for k in range(no_features):
            temp2.append((c[j][k] - features[k][i])**2)
        temps = sum(temp2)
        f.append(temps)
            #f.append(((c[j][0] - x[i])**2 + (c[j][1] - y[i])**2) * 0.5)

    minimum = min_index(f)
    
    temp = [features[j][i] for j in range(no_features)]
    cluster[minimum].append(temp)
    for j in range(no_features):
        c[minimum][j] = (c[minimum][j] + features[j][i])/2 

cluster = np.array(cluster)
for i in range(no_clusters):
    cluster[i] = np.array(cluster[i])
```

```python
colours = ["red", "blue", "yellow", "green", "orange", "purple"]
for i in range(no_clusters):
    plt.scatter(cluster[i][:,0], cluster[i][:,1], color=colours[i])
        
    plt.scatter(c[i][0], c[i][1], color=colours[i], marker="x")

plt.show()
```
