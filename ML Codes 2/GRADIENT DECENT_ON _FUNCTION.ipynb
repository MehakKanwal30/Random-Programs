```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
```

```python
f = lambda x: x**2 + 2
```

```python
Noise = 60
points = 90
```

```python
X = np.array([i for i in range(-1 * points, points + 1)])
Y = np.array([f(x) for x in X])
```

```python
df = lambda x: 2*x
```

```python
plt.plot(X, Y)
plt.show()
```

```python
x = 70
lr = 0.001
for z in range(1001):
    ŷ = f(x)
    
    x = x - lr * df(x)
    
    if z%100 == 0:
        plt.plot(X, Y, c = 'black')
        plt.scatter(x, ŷ, c = "Red")
        plt.show()
```
