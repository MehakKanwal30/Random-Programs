```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand
```

```python
Noise = 100
points  = 20
```

```python
X = np.array([i*10 + rand.randint(0, Noise) for i in range(points)])
Y = np.array([i*10 + rand.randint(0, Noise) for i in range(points)])
```

```python
W = 0
B = 0
lr = 0.000001
for z in range(1001):
     = np.array([W * x  + B for x in X])
    
    dw = (1/len(X)) * np.dot(X.T , ( - Y))
    db = (1/len(X)) * sum( - Y)
    
    W = W - lr * dw
    B = B - lr * db
    
    if z%100 == 0:
        plt.scatter(X, Y)
        plt.plot(X, [W * x  + B for x in X])
        plt.show()
```
